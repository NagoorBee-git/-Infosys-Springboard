```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("INFY_stock_data.csv")
df.head()
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
# Remove commas if present
df.replace(',', '', regex=True, inplace=True)

# Convert columns to numeric
cols = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df['Date'].fillna(method='ffill', inplace=True)

print("Missing values after filling with MEAN:")
print(df.isnull().sum())
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
df['Date'].fillna(method='ffill', inplace=True)

print("Missing values after filling with MEDIAN:")
print(df.isnull().sum())
```

```python
df.fillna(method='ffill', inplace=True)
print("Missing values after filling with FORWARD FILL:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

print("Closing Price Trend")

plt.figure()
plt.plot(df['Date'], df['Close'])
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Closing Price Trend")
plt.show()
```

```python
print("Trading Volume")

plt.figure()
plt.plot(df['Date'], df['Volume'])
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Trading Volume")
plt.show()
```

```python
# Calculate moving averages
df['MA50'] = df['Close'].rolling(50).mean()
df['MA200'] = df['Close'].rolling(200).mean()

print("Moving Average Analysis")

plt.figure()
plt.plot(df['Date'], df['Close'], label="Close Price")
plt.plot(df['Date'], df['MA50'], label="50 Day Average")
plt.plot(df['Date'], df['MA200'], label="200 Day Average")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Moving Average Analysis")
plt.legend()
plt.show()
```

```python
import numpy as np

# Step 1: Select numeric columns automatically
numeric_cols = df.select_dtypes(include=np.number).columns

print("Numeric columns:")
print(numeric_cols)

# Step 2: Calculate Q1, Q3 and IQR
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Step 3: Detect outliers count
outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) |
            (df[numeric_cols] > (Q3 + 1.5 * IQR)))

print("\nOutliers count in each column:")
print(outliers.sum())

# Step 4: Handle outliers using capping (recommended for stock data)
for col in numeric_cols:
    lower = Q1[col] - 1.5 * IQR[col]
    upper = Q3[col] + 1.5 * IQR[col]
    df[col] = np.clip(df[col], lower, upper)

print("\nOutliers handled successfully")

# Step 5: Verify again
outliers_after = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) |
                  (df[numeric_cols] > (Q3 + 1.5 * IQR)))

print("\nOutliers after handling:")
print(outliers_after.sum())
```

```python
import matplotlib.pyplot as plt

df[numeric_cols].boxplot()
plt.title("Boxplot after handling outliers")
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


df['Daily_Return'] = df['Close'].pct_change()
df['MA_10'] = df['Close'].rolling(window=10).mean()
df['MA_20'] = df['Close'].rolling(window=20).mean()
df['Volatility'] = df['Close'].rolling(window=10).std()
```

```python
import numpy as np

df['Target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
features = ['Open', 'High', 'Low', 'Close', 'Volume',
            'Daily_Return', 'MA_10', 'MA_20', 'Volatility']
```

```python
scaler = MinMaxScaler()

df_scaled = df.copy()

df_scaled[features] = scaler.fit_transform(df[features])
```

```python
print("\n===== ORIGINAL DATA (FIRST 5 ROWS) =====")
print(df[features].head())

print("\n===== SCALED DATA (FIRST 5 ROWS) =====")
print(df_scaled[features].head())

print("\n===== FINAL PREPROCESSED DATASET =====")
print(df_scaled.head())
```

```python
df_scaled.to_csv("preprocessed_stock_data.csv", index=False)
```

```python
features = ['Open', 'High', 'Low', 'Close', 'Volume',
            'Daily_Return', 'MA_10', 'MA_20', 'Volatility']

X = df_scaled[features]
y = df_scaled['Target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    shuffle=False
)

print("Training Size:", X_train.shape)
print("Testing Size:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, predictions))
print("\nClassification Report:\n", classification_report(y_test, predictions))
```
